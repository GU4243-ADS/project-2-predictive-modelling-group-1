```python
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import cv2
from os import listdir
from os.path import isfile, join
import pandas as pd
```

## Part 1: Data Cleaning

```python
#load the images into a numpy array.
#inspired by https://stackoverflow.com/questions/33369832/read-multiple-images-on-a-folder-in-opencv-python
mypath = "C:\\Users\\David\\Documents\\train\\" #change this path to wherever your image data is
# the below is a list of individual file names
justfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
#prepare my list of images
images = np.empty(len(justfiles), dtype=object)
#fill the list of images with the dog and cat images
for n in range(0, 500):#len(justfiles)):
  images[n] = cv2.imread( join(mypath,justfiles[n]) )
```

```python
#create an array of images to play around with
tempimg = np.empty(500, dtype=object)
for i in range(500):
    tempimg[i] = images[i]
```

```python
#load the labels, and prepare an array for a bool of labels, dog == True, cat == False
labels = np.loadtxt("..\\data\\train_label.txt", dtype = 'str')
labelsb = np.empty(len(labels), dtype=bool)
```

```python
#make labels a better string
for i in range(len(labels)):
    if ('d' in labels[i]):
        labels[i] = 'dog'
    else:
        labels[i] = 'cat'
```

```python
#fill labelsb
for i in range(len(labels)):
    if ('d' in labels[i]):
        labelsb[i] = True
    else:
        labelsb[i] = False
```

```python
# a small array for testing my ideas
tmp_labelsb = labelsb[range(500)]
```

```python
plt.imshow(gu_tmpimg[30])
plt.show()
```

```python
#make all images gray
g_tmpimg = np.empty(len(tempimg), dtype = object)
for i in range(len(tempimg)):
    g_tmpimg[i] = cv2.cvtColor(tempimg[i], cv2.COLOR_BGR2GRAY)
```

```python
#resize all images; make them uniform
gu_tmpimg = np.empty(len(g_tmpimg), dtype = object)
for i in range(len(tempimg)):
    gu_tmpimg[i] = cv2.resize(g_tmpimg[i], (128,128))
```

```python
# tmp train and test sets
tmp_trn = gu_tmpimg[0:250]
tmp_tst = gu_tmpimg[251:500]
# tmp train and test labels
tmp_ltrn = tmp_labelsb[0:250]
tmp_ltst = tmp_labelsb[251:500]
```

```python
# Initiate STAR detector for all the images
orbs = np.empty(len(tmp_trn), dtype = object)
for i in range(len(tmp_trn)):
    orbs[i] = cv2.ORB_create()

# find the keypoints with ORB
kps = np.empty(len(tmp_trn), dtype=object)
for i in range(len(orbs)):
    kps[i] = orbs[i].detect(tmp_trn[i],None)

# compute the descriptors with ORB
descrs = np.empty(len(kps), dtype=object)
for i in range(len(kps)):
    kps[i],descrs[i] = orbs[i].compute(tmp_trn[i],kps[i])
```

```python
#make descrs all the same dimension
udescrs = np.empty(len(descrs), dtype=object)
for i in range(len(descrs)):
    udescrs[i] = descrs[i][:130]
```

```python
#flatten the arrays
fdescrs = np.empty(len(udescrs), dtype=object)
for i in range(len(udescrs)):
    fdescrs[i] = udescrs[i].flatten()

ftmp_trn = np.empty(len(tmp_trn),dtype=object)
for i in range(len(tmp_trn)):
    ftmp_trn[i] = tmp_trn[i].flatten()

ftmp_tst = np.empty(len(tmp_tst),dtype=object)
for i in range(len(tmp_tst)):
    ftmp_tst[i] = tmp_tst[i].flatten()
```

```python
#reformat the arrays
ftmp_trn = np.vstack(ftmp_trn)
ftmp_tst = np.vstack(ftmp_tst)
fdescrs = np.vstack(fdescrs)
```

```python
#fit the logistic reggresion modelmodel
from sklearn import linear_model
lrmodel = linear_model.LogisticRegression(solver='sag')
lrmodel.fit(ftmp_trn,tmp_ltrn)
```

```python
1 - (sum(lrmodel.predict(ftmp_tst) ^ tmp_ltst)/249)
```

```python
#fit a gradient boosting algorithm
from sklearn import ensemble
gbcmodel = ensemble.GradientBoostingClassifier()
gbcmodel.fit(ftmp_trn,tmp_ltrn)
```

```python
1 - (sum(gbcmodel.predict(ftmp_tst) ^ tmp_ltst)/249)
```

```python
#fit adaboost
adbmodel = ensemble.AdaBoostClassifier()
adbmodel.fit(ftmp_trn,tmp_ltrn)
```

```python
1 - (sum(adbmodel.predict(ftmp_tst) ^ tmp_ltst)/249)
```

```python
#fit svm
from sklearn import svm
svmmodel = svm.SVC()
svmmodel.fit(ftmp_trn,tmp_ltrn)
```

```python
1 - (sum(svmmodel.predict(ftmp_tst) ^ tmp_ltst)/249)
```
